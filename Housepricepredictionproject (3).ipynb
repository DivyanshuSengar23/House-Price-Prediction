## to check the null values by heat plot (True=1, False=0)

```python
plt.figure(figsize=(15,15))
sns.heatmap(df.isnull())
```

## to check null values count in each column

```python
df.isnull().sum()
```

## to check null value percentage in each column

```python
df_null_clm=df.isnull().sum()/df.shape[0]*100
```

## to find the columns which have more the 20% null values

```python
df_null_clm[df_null_clm>20]
```

## to find the columns name which have more the 20% null values

```python
null_clm_list=df_null_clm[df_null_clm>20].index
null_clm_list
```

## to drop the columns which have more the 20% null values

```python
df_drop_clm=df.drop(columns=null_clm_list)
```

```python
df_drop_clm.shape
```

## to select numerical columns 

```python
df_num_clm=df_drop_clm.select_dtypes(include=['int','float'])
```

## to select categorical columns

```python
df_cat_clm=df_drop_clm.select_dtypes(include=['object'])
```

```python
df_cat_clm.columns
```

## filling Numerical missing values 



```python


for col in df_num_clm.columns:
    df_drop_clm[col].fillna(df_drop_clm[col].median(), inplace=True)
```

## filling categorical missing values

```python
for col in df_cat_clm.columns:
    df_drop_clm[col].fillna(df_drop_clm[col].mode()[0], inplace=True)
```

```python
df_drop_clm.isnull().sum().sum()
```

## Encoding

```python
df_encoded = pd.get_dummies(df_drop_clm, drop_first=True)
```

```python
df_encoded.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df_encoded.drop('SalePrice', axis=1)
y = df_encoded['SalePrice']
```

```python

X_train, X_test, y_train, y_test = train_test_split( X, y,test_size=0.2, random_state=12)
```

## Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
df_encoded.shape
```

```python

df_encoded.isnull().sum().sum()
```

```python
df_encoded.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train , y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(X_test,y_test)
```

```python
lr.score(X_test, y_test)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(y_test,y_pred)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual vs Predicted")
plt.show()
```
